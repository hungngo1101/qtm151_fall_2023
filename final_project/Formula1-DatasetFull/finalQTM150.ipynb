```python
import pandas as pd
import numpy as np
```

```python
driver = pd.read_csv('raw_data/drivers.csv')
pitstop = pd.read_csv('raw_data/pit_stops.csv')
result = pd.read_csv ('raw_data/results.csv')
```

```python
driver.describe()
driver
pitstop.describe()
result.describe()
```

```python
pitstop_result = pd.merge(pitstop, result, on = ['driverId', 'raceId'], how = 'outer')
pitstop_result
```

```python
sum_pitstoptime_for_driverId = pitstop_result.groupby('driverId')['milliseconds_x'].sum()
print(sum_pitstoptime_for_driverId)
```

```python
print(sum_pitstoptime_for_driverId.isnull().any())
```
